#### Read the city name from the GitHub link and then store the realtime temperature from google search in sqlite3 DB.

```python
####### Read the city name from the GitHub link and then store the realtime temperature from google search in sqlite3 DB.

import requests
from lxml import html
import time
import sqlite3

conn = sqlite3.connect('Cities_And_Temperature.db')
cursor = conn.cursor()
table = cursor.execute('CREATE TABLE IF NOT EXISTS Cities_Temperature([City Name] text, Temperature text);')

class CitiesTemperature:
	def __init__(self):
		self.google_search_url = "https://www.google.co.in/search"
		self.headers = {
						'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
					   }
		self.count = 0

	def get_Cities_List(self):			### Get all city names from given Github Url and append into a list.
		github_cities_url = "https://github.com/nshntarora/Indian-Cities-JSON/blob/dbae5d894564490a239fff7435841dc120f6ce9e/cities.json"
		tree = html.fromstring(requests.get(github_cities_url).text)
		cities_list = tree.xpath("//*[contains(text(),'name')]//following-sibling::span/text()")
		print("Total Number Of Cities: %s" % len(cities_list))
		return cities_list

	def get_City_Temp(self,cities_list):
		for city in cities_list:
			self.count += 1
			params = {
						"q": f"{city} temperature"
					 }
			try:
				response = requests.get(self.google_search_url, headers=self.headers, params=params, timeout=10)
				tree = html.fromstring(response.text)
				temperature = tree.xpath('//*[@id="wob_tm"]//text()')[0]+"°C"
				query = ("Insert into Cities_Temperature ('City Name','Temperature') values {}".format(tuple([city,temperature])))
				cursor.execute(query)
				conn.commit()
				print("%s Temperature of %s %s" % (self.count, city,temperature))
			except:
				print("%s Temperature couldn't be found %s" % (self.count, city))
			time.sleep(1)
		conn.close()

if __name__ == '__main__':
	obj = CitiesTemperature()
	cities_list = obj.get_Cities_List()
	obj.get_City_Temp(cities_list)
```
